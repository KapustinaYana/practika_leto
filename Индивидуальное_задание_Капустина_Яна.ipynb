## 3. Фильтрация данных

```python
filtered_sbor = marvel.query('opening_weekend_us > 70000000') #Фильтрация с помощью query
print('Фильтрация фильмов с открывающими сборами больше 70 миллионов')
print(filtered_sbor)
```

```python
filtered_length = marvel.query('length_hrs > 1')
print('Фильтрация фильмов длиной больше 1 часа')
print(filtered_length)
```

```python
filtered_gross_high = marvel.query('gross_world > 400000000')
print('Фильтрация фильмов с общими сборами более 400 миллионов')
print(filtered_gross_high)
```

```python
filtered_film = marvel.query('length_hrs > 2 and gross_world > 300000000')
print('Фильтрация фильмов, которые длиннее 2 часов и имеют кассовые сборы более 300 миллионов')
print(filtered_film)
```

```python
filtered_sbor60 = marvel.query('opening_weekend_us < 60000000')
print('Фильтрация фильмов, которые имели кассовые сборы за первый показ менее 60 миллионов')
print(filtered_sbor60)
```

```python
filtered_opening_where = marvel.where((marvel['opening_weekend_us'] > 70000000) & (marvel['opening_weekend_us'] < 100000000)).dropna() #Фильтрация с помощью where
print('Фильтрация фильмов с открывающим сбором среди 70 - 100 миллионов')
print(filtered_opening_where)
```

```python
filtered_length_where = marvel.where(marvel['length_hrs'] < 2).dropna()
print('Фильтрация фильмов с длительностью менее 2 часов')
print(filtered_length_where)
```

```python
filtered_gross_where = marvel.where((marvel['gross_world'] > 500000000) & (marvel['gross_world'] < 800000000)).dropna()
print('Фильтрация по мировой кассе между 500 и 800 миллионами')
print(filtered_gross_where)
```

```python
filtered_film_USA_where = marvel.where(marvel['gross_us'] < 200000000).dropna()
print('Фильтрация фильмов, которые собрали менее 200 миллионов долларов в США')
print(filtered_film_USA_where)
```

```python
filtered_film_world_where = marvel.where(marvel['gross_world'] < 400000000).dropna()
print('Фильтрация фильмов, которые собрали менее 400 миллионов долларов в мире')
print(filtered_film_world_where)
```

## 4. Сводные таблицы

```python
pivot__sbor_USA_gross_us = marvel.pivot_table(values = 'gross_us', index = 'length_min', aggfunc = 'mean')
print('Сводная таблица по средней сумме кассовых сборов в США по длительности фильма')
print(pivot__sbor_USA_gross_us)
```

```python
pivot_total_gross_world = marvel.pivot_table(values = 'gross_world', index = 'length_min', aggfunc = 'sum')
print('Сводная таблица по общей сумме мировых сборов по длительности фильма')
print(pivot_total_gross_world)
```

```python
opening_weekend_summary = marvel.pivot_table(values = 'opening_weekend_us', index = 'movie', aggfunc = ['min', 'max', 'mean'])
print('Сводная таблица по открывающим сборам')
print(opening_weekend_summary)
```

## 5. Группировка данных, агрегатные функции

```python
marvel[['length_hrs', 'length_min', 'opening_weekend_us', 'gross_us', 'gross_world']].agg(['min', 'max', 'mean', 'median', 'sum'])
```

```python
grupirovka = {
    'length_hrs': ['min', 'max', 'mean', 'median'],
    'length_min': ['min', 'max', 'mean', 'median'],
   'opening_weekend_us': ['min', 'max', 'mean', 'median'],
    'gross_us': ['min', 'max', 'mean', 'median'],
    'gross_world': ['min', 'max', 'mean', 'median'],
}
agregat_grupirovka = marvel.groupby('movie').agg(grupirovka).reset_index()
print(agregat_grupirovka.head(5))
```

## 6. Поиск закономерностей, варианты решения бизнес-задачи

```python
marvel
```

```python
#Группировка по названию и вычисление средних открывающих и общих кассовых сборов
grouped = marvel.groupby('movie').agg({
    'opening_weekend_us': 'mean',
    'gross_us': 'mean',
    'gross_world': 'mean'
}).reset_index()
gross_us_Usa = grouped.sort_values(by = 'gross_us', ascending = False) #Сортировка по средним кассовым сборам в США
print(gross_us_Usa)
```

# Графический анализ данных

## 1. Диаграммы, библиотека Matplotlib

```python
marvel
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(data = marvel, x = 'length_hrs', bins = 10, kde = True)
plt.title('Распределение продолжительности фильмов', fontsize = 15)
plt.xlabel('Длительность (в часах)', fontsize = 10)
plt.ylabel('Частота', fontsize = 10 )
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(marvel['length_hrs'], marvel['gross_us'], alpha = 0.7)
plt.title('Сборы фильмов в США в зависимости от продолжительности фильма', fontsize = 15)
plt.xlabel('Длительность (в часах)', fontsize = 10)
plt.ylabel('Сборы в США (доллары)', fontsize = 10)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(marvel['opening_weekend_us'], marvel['gross_world'], alpha = 0.7)
plt.title('Сборы фильмов в мире в зависимости от кассовых сборов за первый показ', fontsize = 15)
plt.xlabel('Сбор за первый показ', fontsize = 10)
plt.ylabel('Сборы в мире (доллары)', fontsize = 10)
plt.grid(True)
plt.show()
```

## 2. Диаграммы, библиотека Seaborn

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data = marvel, x='movie', y='gross_world', marker='o')
plt.title('Рост сборов в мировом прокате по годам', fontsize = 15)
plt.xlabel('Название фильма', fontsize = 10)
plt.ylabel('Сборы в мире', fontsize = 10)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sbor = marvel['movie'].value_counts().head(5)
sns.barplot(x = sbor.values, y = sbor.index)
plt.title('Топ 5 фильмов Marvel, которые нравятся людям')
plt.xlabel('Сколько денег получили за фильм')
plt.ylabel('Название фильма')
plt.grid(axis = "x", linestyle = "--")
plt.show()
```
